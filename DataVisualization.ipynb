```python
# 1.29 (c)

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

df = pd.DataFrame({'Religions': [376,2100,900,1500,400,14,23,394,61]})
df.index = ["Buddhism", "Christianity", "Hinduism", "Islam", 
            "Primal indigenous and African traditional", "Judaism", "Sikhism", "Chinese traditional", "Other"]
df = df.sort_values(by='Religions',ascending=True)
df["cumpercentage"] = df["Religions"].cumsum()/df["Religions"].sum()*100


fig, ax = plt.subplots()
ax.bar(df.index, df["Religions"], color="C0")
ax2 = ax.twinx()
ax2.plot(df.index, df["cumpercentage"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
plt.show()
```
